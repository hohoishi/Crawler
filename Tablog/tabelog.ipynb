```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup
import pandas as pd 


# response = urlopen(url)  #這個是檔案型態
#json.load() -> 把檔案型態(支援.read())轉換成list/dict型態
# html = BeautifulSoup(response)  #用湯轉換成html 
# print(html)
#區塊型態.find/find_all(html名字, class/id篩選)
# target = html.find('a',{'class':'list-rst__rst-name-target'})
#區塊:
#1. 特殊屬性:區塊['href']
#2. html內容:區塊.text
#一個型態可以有兩種東西 : a. 專屬功能:人.吃飯() b. 專屬值: 人.身高
# print(target['href'])
# print(target.text)
dict = {
    'name':[],
    'title':[],
    'rating':[],
    'budget_dinner':[],
    'budget_lunch':[]
}
page = 57
while True:
    print(f'頁數: {page}')
    url = f'https://tabelog.com/tw/tokyo/rstLst/{page}/'
    page = page+1
    response = urlopen(url)
    html = BeautifulSoup(response)
    rs = html.find_all('div', {'class':'list-rst__rst-data'})
    if rs:
        for r in rs :
            en = r.find('a', {'class':'list-rst__rst-name-target'})
            title = r.find('div', {'class':"list-rst__area-genre"})
            p = r.find('p', {'class':'list-rst__pr-title'})
            rating = r.find('span', {'class':'c-rating__val'})
            budget = r.find_all('span', {'class':'c-rating-v3__val'})

            print(en.text)
            print(title.text.strip())
            # print(p.text.strip())
            print(rating.text)
            print(en['href'])
            print(f'晚餐預算: {budget[0].text}\n午餐預算: {budget[1].text}')
            print('-'*30)

            dict['name'].append(en.text)
            dict['title'].append(title.text.strip())
            dict['rating'].append(rating.text)
            dict['budget_dinner'].append(budget[0].text)
            dict['budget_lunch'].append(budget[1].text)
    else:
        break
df = pd.DataFrame(dict)
df.to_csv('tabelog.csv', encoding='utf-8', index=False)    

```
