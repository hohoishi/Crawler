```python
import requests
import json
import pandas as pd 
payloads={
    'sn': '23505'
}
url = 'https://ani.gamer.com.tw/ajax/danmuGet.php'
header = {'User-Agent':'xxxxx'}
response = requests.post(url ,payloads, headers=header)
danmu = json.loads(response.text)

dic = {
    'userid':[],
    'text':[]
}
for d in danmu:
    print(d['userid'], d['text'])
    dic['userid'].append(d['userid'])
    dic['text'].append(d['text'])

df = pd.DataFrame(dic)
df.to_csv('baha_danmu.csv',encoding='utf-8',index=False)
```
